```python
# Exercise 1: Basic Math Operations
# i. Addition
a = 5
b = 3
print("Sum:", a + b)

# ii. Subtraction
print("Difference:", a - b)

# iii. Multiplication
print("Product:", a * b)

# iv. Division
print("Quotient:", a / b)

# v. Modulus (Remainder)
print("Remainder:", a % b)

# vi. Exponentiation (Power)
print("Power:", a ** b)

# Exercise 2: Python Data Types
# i. List of fruits
fruits = ["Apple", "Banana", "Cherry", "Date", "Orange"]
print("Fruits:", fruits)

# ii. Tuple of animals
animals = ("Dog", "Cat", "Cow")
print("Animals:", animals)

# iii. Dictionary of programming languages
languages = {"Python": 1991, "Java": 1995, "C++": 1985}
print("Programming Languages:", languages)

# iv. Set of colors
colors = {"Red", "Green", "Blue"}
print("Colors:", colors)

# v. String variable
sentence = "skibidi sigma"
print("Sentence:", sentence)

# Exercise 3: Conditions
# i. Check if number is positive, negative, or zero
num = int(input("Enter a number: "))
if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

# ii. Check if string starts with a vowel or consonant
text = input("Enter a string: ")
if text[0].lower() in "aeiou":
    print("The string starts with a vowel.")
else:
    print("The string starts with a consonant.")

# iii. Check if two numbers are equal or not
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
if x == y:
    print("The numbers are equal.")
else:
    print("The numbers are not equal.")

# Exercise 4: Loops
# i. Print numbers from 1 to 10 using a for loop
for i in range(1, 11):
    print(i)

# ii. Print even numbers from 2 to 10 using a for loop
for i in range(2, 11, 2):
    print(i)

# iii. Print the first 5 Fibonacci numbers using a while loop
fib = [0, 1]
while len(fib) < 5:
    fib.append(fib[-1] + fib[-2])
print("Fibonacci sequence:", fib)

# Exercise 5: Functions
# i. Function to add two numbers
def add_numbers(x, y):
    return x + y

# ii. Function to find the largest number in a list
def find_largest(numbers):
    return max(numbers)

# iii. Function to count vowels in a string
def count_vowels(s):
    return sum(1 for char in s.lower() if char in "aeiou")

# Testing the functions
print("Sum of 5 and 3:", add_numbers(5, 3))
print("Largest number in [10, 20, 5, 8]:", find_largest([10, 20, 5, 8]))
print("Number of vowels in 'Hello World':", count_vowels("Hello World"))
```
